```python
df.columns = df.columns.str.strip().str.lower()
df.columns
```

```python
df = df.dropna(subset=["size"])
```

```python
df["podate"] = pd.to_datetime(df["podate"])
```

```python
df["receivingdate"] = pd.to_datetime(df["receivingdate"])
```

```python
 df["invoicedate"] = pd.to_datetime(df["invoicedate"])
```

```python
df["paydate"] = pd.to_datetime(df["paydate"])
```

```python
df["lead_time"] = (df["receivingdate"] - df["podate"]).dt.days
```

```python
df["payment_delay"] = (df["paydate"] - df["invoicedate"]).dt.days
```

```python
top_vendors = df.groupby("vendorname")["dollars"].sum().sort_values(ascending=False).head(10)
print(top_vendors)
```

```python
# Group by Description to see which products we buy the most of
top_qty_brands = df.groupby("description")["quantity"].sum().sort_values(ascending=False).head(10)
print(top_qty_brands)
```

```python
vendor_lead_time = df.groupby("vendorname")["lead_time"].mean().sort_values(ascending=False).head(10)
print(vendor_lead_time)
```

```python
import matplotlib.pyplot as plt

# 1. Create the data (reuse the logic from earlier)
top_vendors = df.groupby("vendorname")["dollars"].sum().sort_values(ascending=False).head(10)

# 2. Create the plot
plt.figure(figsize=(12,6)) # Makes the chart wide
top_vendors.plot(kind='bar', color='red')

# 3. Add Labels (This makes it professional)
plt.title('Top 10 Vendors by Total Spend', fontsize=15)
plt.xlabel('Vendor Name', fontsize=12)
plt.ylabel('Total Dollars (Millions)', fontsize=12)
plt.xticks(rotation=90, ha='right') # Tilts the names so they don't overlap

# 4. Show it
plt.show()
```

```python
# 1. Get the average lead time for the top 10 vendors we buy from most often
top_10_vendors = df['vendorname'].value_counts().head(10).index
vendor_efficiency = df[df['vendorname'].isin(top_10_vendors)].groupby('vendorname')['lead_time'].mean().sort_values()

# 2. Plotting
plt.figure(figsize=(10,6))
vendor_efficiency.plot(kind='barh', color='green') # 'barh' makes it a horizontal bar chart

# 3. Aesthetics
plt.title('Average Lead Time: Top 10 Frequent Vendors', fontsize=14)
plt.xlabel('Days to Receive Order', fontsize=12)
plt.ylabel('Vendor Name', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7) # Adds vertical dashed lines for easier reading

plt.show()
```

```python
# 1. Set the date as the index (required for time analysis)
df_time = df.set_index('podate')

# 2. Group all data by Month ('M') and sum the Dollars
monthly_spend = df_time['dollars'].resample('ME').sum()

# 3. Create a Line Chart
plt.figure(figsize=(12,5))
plt.plot(monthly_spend.index, monthly_spend.values, marker='o', linestyle='-', color='green')

# 4. Labels
plt.title('Total Purchase Spend by Month (2016)', fontsize=15)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Spend (Millions)', fontsize=12)
plt.grid(True, alpha=0.3)

plt.show()
```

```python
# 1. Prepare the monthly data again
df_time = df.set_index('podate')
monthly_spend = df_time['dollars'].resample('ME').sum()

# 2. Start the Plot
plt.figure(figsize=(14,7))
plt.plot(monthly_spend.index, monthly_spend.values, marker='o', linestyle='-', color='teal', linewidth=2)

# 3. Add Annotations (The "Consultant" Touch)
# We use .annotate() to point to specific dates

# Marking August (Back to School / Summer End)
plt.annotate('Back to School / Summer Peak', 
             xy=(monthly_spend.index[7], monthly_spend.values[7]), 
             xytext=(monthly_spend.index[5], monthly_spend.values[7]*1.1),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, fontweight='bold')

# Marking November (Thanksgiving / Holiday Prep)
plt.annotate('Thanksgiving & Holiday Stock-up', 
             xy=(monthly_spend.index[10], monthly_spend.values[10]), 
             xytext=(monthly_spend.index[8], monthly_spend.values[10]*1.1),
             arrowprops=dict(facecolor='red', shrink=0.05),
             fontsize=10, fontweight='bold', color='red')

# 4. Labels and Styling
plt.title('2016 Purchasing Trends with Key Business Drivers', fontsize=16)
plt.ylabel('Total Spend (Dollars)', fontsize=12)
plt.grid(True, alpha=0.2)

plt.show()
```

```python
# 1. Calculate the 'Space-to-Value' ratio
# We want to see which items have high quantity but low dollar impact
df['qty_per_dollar'] = df['quantity'] / df['dollars']

# 2. Find the top 10 "Space Wasters" (High quantity, low price)
dead_stock_candidates = df.groupby('description').agg({
    'quantity': 'sum',
    'dollars': 'sum',
    'qty_per_dollar': 'mean'
}).sort_values(by='qty_per_dollar', ascending=False).head(10)

print(dead_stock_candidates)
```

```python
# Find all rows where quantity is greater than 0 but dollars are 0
zero_dollar_orders = df[(df['quantity'] > 0) & (df['dollars'] == 0)]

# See how many rows have this issue
print(f"Total Zero-Dollar Rows: {len(zero_dollar_orders)}")

# Look at the top 5 vendors who are 'giving away' free stuff
print(zero_dollar_orders['vendorname'].value_counts().head())
```

```python
# 1. Calculate the average price for Macallan 12 from the REST of the data
macallan_avg_price = df[df['description'] == 'The Macallan Double Cask 12']['purchaseprice'].mean()

# 2. Estimate the missing value for just that one product
missing_value = 2015 * macallan_avg_price

print(f"Estimated Missing Value for Macallan 12: ${missing_value:,.2f}")
```

```python
# 1. Get the price using the Brand ID (2520)
price_2520 = df[df['brand'] == 2520]['purchaseprice'].mean()
print(f"Price for Brand 2520: ${price_2520:.2f}")

# 2. Re-calculate the estimated value for those 2,015 bottles
final_missing_value = 2015 * price_2520
print(f"Final Corrected Estimated Value: ${final_missing_value:,.2f}")
```
